```python
import numpy as np
import matplotlib.pyplot as plt

greyhounds = 500
labs = 500

grey_height = 28 + 4 * np.random.randn(greyhounds)
lab_height = 24 + 4 * np.random.randn(labs)

plt.hist([grey_height, lab_height], stacked=True, color=['r','b'])
plt.show()
```

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
stock_df = pd.read_csv('data/S&P500_Stock_Data.csv')
stock_df
```

```python
sns.pairplot(stock_df)
```

```python
stock_df.describe().round(2)
```

```python
from sklearn.preprocessing import MinMaxScaler
stock_df = stock_df.drop('Date', axis=1)
scaler = MinMaxScaler()
stock_df = scaler.fit_transform(stock_df)
```

```python
stock_df = pd.DataFrame(stock_df, columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Dividends', 'Stock Splits'])
```
